```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
year = df.drop('per capita income (US$)',axis='columns')
year
```

```python
income = df["per capita income (US$)"]
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(year,income,color='red',marker='+')  
plt.plot(year,reg.predict(year),color="blue")
```
